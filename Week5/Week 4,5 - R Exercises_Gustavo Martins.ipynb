3. Input the peanut_lines (CSV) and peanut_disease (Tab) datafile

```r
peanut_lines<-read.csv(file = "peanut_lines.csv",header = T)
head(peanut_lines)
```

```r
peanut_disease<-read.table("peanut_disease.txt",header=T,sep="\t")
head(peanut_disease)
```

4. Check the column names of peanut_lines and peanut_disease

```r
colnames(peanut_lines)
str(peanut_lines)
write.csv(peanut_lines,"peanut_lines_1.csv")
```

```r
colnames(peanut_disease)
str(peanut_disease)
write.csv(peanut_disease,"peanut_disease_1.csv")
```

5. Merge peanut_lines and peanut_disease keeping the entirety of the peanut_disease information intact - Name the data frame disease_data

```r
disease_data<-merge(peanut_disease,peanut_lines,by="NC_Accession",all=F,all.x = T,all.y=F)
str(disease_data)
write.csv(disease_data,"disease_data_1.csv")
```

6. Subset the disease data to list only ratings for leaf spot ('LS') in a dataframe called leaf.spot

```r
leaf.spot<-disease_data[disease_data[,"Rating"]=="LS",]
leaf.spot
```

7. Drop all rows that contain missing values in the column 'Disease'

```r
leaf.spot[-(which(is.na(leaf.spot[,"Disease"]))),]
```

8. What was the average leaf spot rating in 2017?

```r
paste("The average leaf spot rating in 2017 was",round(mean(leaf.spot[which(leaf.spot[,"Year"]==2017),"Disease"],na.rm=T),2))
```
